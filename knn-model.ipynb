```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import math
from sklearn.metrics import confusion_matrix
```

```python
dataset = pd.read_csv("/kaggle/input/dataset/KNN_Dataset.csv")
dataset.head()
```

```python
zero_not_accepted = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
```

```python
for column in zero_not_accepted:
    dataset[column] = dataset[column].replace(0, np.NaN)
    mean = dataset[column].mean(skipna=True)
    dataset[column] = dataset[column].replace(np.NaN, mean)

dataset.head()
```

```python
x = dataset.iloc[:, 0:8]
y = dataset.iloc[:, 8]

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.2)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
k = round(math.sqrt(len(y_test)))
model = KNeighborsClassifier(k, p=2, metric="euclidean")
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
con_mat = confusion_matrix(y_test, y_predict)
total_accuracy = (con_mat[0, 0] + con_mat[1, 1]) / float(np.sum(con_mat))
class1_accuracy = (con_mat[0, 0] / float(np.sum(con_mat[0, :])))
class2_accuracy = (con_mat[1, 1] / float(np.sum(con_mat[1, :])))

print('Total accuracy:\t\t%.2f' % total_accuracy)
print('Class1 accuracy:\t%.2f' % class1_accuracy)
print('Class2 accuracy:\t%.2f' % class2_accuracy)
print('Geo. mean accuracy:\t%.2f' % math.sqrt((class1_accuracy * class2_accuracy)))
```
